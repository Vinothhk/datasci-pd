```python
import statistics as st
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diabetes = pd.read_csv('../datasets/diabetes.csv')
```

```python
age_mean = st.mean(diabetes['Age'])
age_mean
```

```python
bp_stddev = st.stdev(diabetes['BloodPressure'])
bp_stddev
```

```python
gluc_median = st.median(diabetes['Glucose'])
gluc_median
```

```python
DiaPedFunc_var = st.variance(diabetes['DiabetesPedigreeFunction'])
DiaPedFunc_var
```
