```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics as st
```

```python
%matplotlib inline
%matplotlib widget
```

```python
dataset = pd.read_csv('newDataset.csv')
dataset
```

```python
tool_pos = dataset[['x','y','z']]
# print(tool_pos.iloc[1,2])
tool_pos
```

```python
tool_x = dataset['x']
tool_y = dataset['y']
tool_z = dataset['z']
ax = plt.axes(projection='3d')
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.set_zlim(-1,1)
plt.scatter(tool_x,tool_y,tool_z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.tight_layout()
plt.show()    
```

```python
print(st.mean(tool_x))
print(st.stdev(tool_x))
```

```python
print(st.mean(tool_y))
print(st.stdev(tool_y))
```

```python
print(st.mean(tool_z))
print(st.stdev(tool_z))
```
