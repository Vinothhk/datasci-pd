```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
%matplotlib widget
```

```python
dataset = pd.read_csv('robot_inverse_kinematics_dataset.csv')
dataset
```

```python
tool_pos = dataset[['x','y','z']]
# print(tool_pos.iloc[1,2])
tool_pos
```

```python
x_ = []
y_ = []
z_ = []
for i in range(0,len(tool_pos)):
    x,y,z = tool_pos.iloc[i,:3]
    x_.append(x)
    y_.append(y)
    z_.append(z)
ax = plt.axes(projection='3d')
plt.scatter(x_,y_,z_)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.tight_layout()
plt.show()    
```
