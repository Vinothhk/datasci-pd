```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
import csv
```

```python
%matplotlib inline
%matplotlib widget
```

```python
x_min, x_max = -0.58, 0.58
y_min, y_max = -0.58, 0.58
z_min, z_max = -0.3, 0.87

# Number of random points to generate
num_points = 120000

# Generate random 3D points within the workspace
x = np.random.uniform(x_min, x_max, num_points)
y = np.random.uniform(y_min, y_max, num_points)
z = np.random.uniform(z_min, z_max, num_points)

# Combine the x, y, z coordinates into a single array
random_points = np.vstack((x, y, z)).T

# Print the generated points
print(random_points)
```

```python
print(x[1],y[1],z[1])
```

```python
st.stdev(x)
```

```python
ax = plt.axes(projection='3d')
ax.plot(x,y,z)
plt.show()
```

```python
with open('WorkspaceDataset120K.csv', mode='w') as datafile:
    writer = csv.writer(datafile, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    writer.writerow(['x','y','z'])
    for i in range(len(x)):
        writer.writerow([x[i],y[i],z[i]])
```
